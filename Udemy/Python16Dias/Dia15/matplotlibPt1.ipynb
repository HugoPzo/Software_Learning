#### Hay otra manera de crear gráficos en Matplotlib, utilizando el método orientado a objetos

```python
# Creamos el gráfico utilizando plt.subplots()
fig, ax = plt.subplots()
# fig - todo el canvas, la figura del grafico
# ax - Eje del grafico

ax.plot(x, y)
```

#### Flujo de creacion de matplotlib

```python
# Importar y preparar la libreria
import matplotlib.pyplot as plt

# Preparar los datos
x = list(range(101))
y = [numero**2 for numero in x]

# Preparamos el area del grafico (fig), y el grafico en si (ax) utilizando plt.subplots()
fig, ax = plt.subplots()

# Añadimos datos al grafico
ax.plot(x, y)

# Personalizamos el grafico, añadiendo titulo y nombre a los ejes
ax.set(title="Casos de COVID-19", xlabel="Dias", ylabel="Casos")

# Guardamos el grafico
fig.savefig("casos.jpg")
```

```python
# Creamos un nuevo set de datos utilizando la libreria de numpy
import numpy as np

# Crea una serie de puntos en una linea establecida
                # comienzo, final, cantidad de puntos
x_1 = np.linspace(0, 100, 20)
y_1 = x_1 ** 2

# Creamos el grafico de dispersion
fig, ax = plt.subplots()
# Punteado
ax.scatter(x_1, y_1)
```

```python
# Visualizamos ahora la funcion seno, utilizando np.sin(x)
fig, ax = plt.subplots()

x_2 = np.linspace(-10, 10, 100)
y_2 = np.sin(x_2)
print(y_2) # Una array de 100 numeros, uno por cada de x_2

ax.scatter(x_2, y_2)
```
