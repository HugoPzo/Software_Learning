#### Flujo de creacion de matplotlib

```python
# Importar y preparar la libreria
import matplotlib.pyplot as plt

# Preparar los datos
x = list(range(101))
y = [numero**2 for numero in x]

# Preparamos el area del grafico (fig), y el grafico en si (ax) utilizando plt.subplots()
fig, ax = plt.subplots()

# Añadimos datos al grafico
ax.plot(x, y, linestyle="--")

# Personalizamos el grafico, añadiendo titulo y nombre a los ejes
ax.set(title="Casos de COVID-19", xlabel="Dias", ylabel="Casos")

# Guardamos el grafico
fig.savefig("casos.jpg")
```

```python
# Creamos un nuevo set de datos utilizando la libreria de numpy
import numpy as np

# Crea una serie de puntos en una linea establecida
                # comienzo, final, cantidad de puntos - En este caso cada 5 - 100\20
x_1 = np.linspace(0, 100, 20)
y_1 = x_1 ** 2


# Creamos el grafico de dispersion
fig, ax = plt.subplots()
# Punteado
ax.scatter(x_1, y_1)
```

```python
# Creemos un diccionario con tres platos y su respectivo precio
# Las claves del diccionario serán los nombres de las comidas, y los valores asociados, su precio
import matplotlib.pyplot as plt

comidas = {'Pizza' : 350, 'Pollo' : 612, 'Tomahawk' : 924}
```

```python
# Crearemos un gráfico de barras donde el eje x está formado por las claves del diccionario,
# y el eje y contiene los valores.
fig, ax = plt.subplots()

# Grafico de barras
ax.bar(comidas.keys(), comidas.values())

# Añadimos los títulos correspondientes
ax.set(title='Comidas - Precios', xlabel='Comidas', ylabel='Precios')
```

```python
# Probemos a continuación con un gráfico de barras horizontales
fig, ax = plt.subplots()

# Grafico de barras horizontal
ax.barh(comidas.keys(), comidas.values())
ax.set(title='Comidas - Precio', xlabel='Precios', ylabel='Comidas')
```

Un gráfico semejante es un histograma. Podemos generar números aleatorios que siguen una distribución normal (que se acumulan en torno a un valor central), con la función randn:

np.random.randn es una función de NumPy que genera números aleatorios siguiendo una distribución normal estándar (media = 0, desviación estándar = 1).

La desviación estándar mide cuánto se dispersan los valores en un conjunto de datos respecto a la media.


```python
# Creamos una distribución de 1000 valores aleatorios distribuidos normalmente
import numpy as np
x = np.random.randn(1000)


# Creamos el histograma
fig, ax = plt.subplots()
ax.hist(x) # La mayoria de los numero se acercaran al 0
# Si generamos muchos números con np.random.randn(), alrededor del 68% estarán entre -1 y 1 (una desviación estándar de la media), y el 95% estarán entre -2 y 2.
```

```python
# Creamos una figura con 4 subgráficos (2 por fila)
# Creamos la misma disposición de gráficos, con un tamaño de figura de 10x5

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))
```

```python
# Creamos la misma disposición de gráficos, con un tamaño de figura de 10x5
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Para nuestro primer gráfico, tomamos el conjunto x_1, y_1, y generamos un gráfico de líneas
ax1.plot(x_1, y_1)

# Para nuestro segundo gráfico, tomamos el conjunto x_2, y_2, y generamos un gráfico de dispersión - np.linspace(0, 100, 20)
x_2 = np.linspace(-10, 10, 100)
y_2 = np.sin(x_2)
ax2.scatter(x_2, y_2)

# Creamos un gráfico con los precios de tres comidas en la esquina inferior izquierda
ax3.barh(comidas.keys(), comidas.values())

# El gráfico de la esquina inferior derecha será un histograma de valores aleatorios con distribución normal
ax4.hist(x)
```
